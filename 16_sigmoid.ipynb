3) Evaluate the output of the xor problem. 

```python
w1 = np.array([[1, -3, 4], [5, 2, -5],[-5, 5,-1]])   # first layer weights top row is the bias
w2 = np.array([[-2.5, 1 , 1, 1]]).T     # second layer weight. Frist element is the bias
print(w1)
print(w2)
```

```python
from scipy.stats import logistic
hactivation = logistic.cdf(np.hstack((np.ones((K,1)), X)).dot(w1),1)
output1 = logistic.cdf(np.hstack((np.ones((K,1)), hactivation)).dot(w2),1)
print(hactivation)
print(output1)
```
